```python
## 5.Entropy Calculation:Given a dataset with 14 instances where 9 are "Yes" and 5 are "No"
## calculate the total Entropy of the system using the formula
## $$H(S) = -\sum_{i=1}^{c} p_i \log_2 p_i$$

# Answers

# Entropy Calculation
import math

# Given data
total = 14
yes = 9
no = 5

# Probabilities
p_yes = yes / total
p_no = no / total

print(f"p_yes = {yes}/{total} = {p_yes:.3f}")
print(f"p_no = {no}/{total} = {p_no:.3f}")

# Entropy calculation
entropy = -(p_yes * math.log2(p_yes) + p_no * math.log2(p_no))

print(f"\nH(S) = -({p_yes:.3f} * log2({p_yes:.3f}) + {p_no:.3f} * log2({p_no:.3f}))")
print(f"H(S) ≈ {entropy:.3f}")
```

```python
## 6.Information Gain:Suppose you split the dataset above on a feature "Weather" which has two outcomes
## Sunny: 3 "Yes", 2 "No",Rainy: 6 "Yes", 3 "No" ,Calculate the Information Gain for the "Weather" feature

# Answers
#=> Information Gain Subset Entropy

# Gain Calculation
import math

def entropy(yes, no):
    total = yes + no
    p_yes = yes / total
    p_no = no / total
    return -(p_yes * math.log2(p_yes) + p_no * math.log2(p_no))

# Given data
total_yes = 9
total_no = 5

# Sunny: 3 Yes, 2 No
sunny_yes, sunny_no = 3, 2
sunny_total = sunny_yes + sunny_no
h_sunny = entropy(sunny_yes, sunny_no)

# Rainy: 6 Yes, 3 No  
rainy_yes, rainy_no = 6, 3
rainy_total = rainy_yes + rainy_no
h_rainy = entropy(rainy_yes, rainy_no)

print(f"Sunny ({sunny_yes} Yes, {sunny_no} No) → H = {h_sunny:.3f}")
print(f"Rainy ({rainy_yes} Yes, {rainy_no} No) → H = {h_rainy:.3f}")

# Weighted entropy
total = sunny_total + rainy_total
weighted_entropy = (sunny_total / total) * h_sunny + (rainy_total / total) * h_rainy

print(f"\nH(S, Weather) = {sunny_total}/{total}({h_sunny:.3f}) + {rainy_total}/{total}({h_rainy:.3f})")
print(f"H(S, Weather) = {weighted_entropy:.3f}")

# Original entropy
original_entropy = entropy(total_yes, total_no)
print(f"\nOriginal H(S) = {original_entropy:.3f}")

# Information gain
gain = original_entropy - weighted_entropy
print(f"\nGain = {original_entropy:.3f} - {weighted_entropy:.3f}")
print(f"Gain = {gain:.3f}")
```

```python
## Gini Impurity:Calculate the Gini Impurity for a node that contains 4 instances of Class A
## 3 instances of Class B, and 3 instances of Class C.$$Gini = 1 - \sum_{i=1}^{c} (p_i)^2$$

# Answers
#=> Gini Impurity Calculate

# Gini Impurity - Simple Calculation

total = 10
A, B, C = 4, 3, 3

p_A = A / total
p_B = B / total
p_C = C / total

gini = 1 - (p_A**2 + p_B**2 + p_C**2)

print(f"Gini = 1 - ({p_A:.2f}² + {p_B:.2f}² + {p_C:.2f}²)")
print(f"Gini = 1 - ({p_A**2:.2f} + {p_B**2:.2f} + {p_C**2:.2f})")
print(f"Gini = 1 - {p_A**2 + p_B**2 + p_C**2:.2f}")
print(f"Gini = {gini:.2f}")
```

```python
## 11.Hyperparameter Tuning: Using a library like Scikit-Learn, load the Iris or Titanic dataset
## Train a Decision Tree and experiment with the following parameters.. Document how each affects the model's accuracy and tree depth
## max_depth => min_samples_split => ccp_alpha (Cost Complexity Pruning)

# Answers
#=> max_depth => Controls depth
#=> min_samples_split => Avoids small splits
#=> ccp_alpha => Prunes tree

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load dataset
data = load_iris()
X = data.data
y = data.target

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = DecisionTreeClassifier(max_depth=3, min_samples_split=4, ccp_alpha=0.01)
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

# Results
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Tree Depth:", model.get_depth())
```
